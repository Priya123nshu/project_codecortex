# Kaggle Low-Latency Avatar Demo

This notebook avoids MuSeTalk in the live loop. It renders a browser-based 2D avatar whose mouth reacts directly to the audio being played or captured.

## How to use
- Leave `SAMPLE_AUDIO_PATH = None` if you only want upload + microphone controls.
- Set `SAMPLE_AUDIO_PATH` to an MP3/WAV path if you want a bundled sample button.
- Run the code cell below. It will save a standalone HTML page into `/kaggle/working/low_latency_avatar_demo.html` and render the same UI inline.
- If microphone permissions are blocked inside the notebook output, download the saved HTML file and open it in a regular browser tab.


```python
from pathlib import Path
from IPython.display import HTML, display

from kaggle_low_latency_avatar import audio_path_to_data_url, build_demo_html

SAMPLE_AUDIO_PATH = None
# Example: SAMPLE_AUDIO_PATH = '/kaggle/input/my-audio-pack/demo.mp3'

sample_audio_data_url = None
if SAMPLE_AUDIO_PATH:
    sample_audio_data_url = audio_path_to_data_url(SAMPLE_AUDIO_PATH)

html = build_demo_html(sample_audio_data_url)
output_dir = Path('/kaggle/working') if Path('/kaggle/working').exists() else Path('.')
output_path = output_dir / 'low_latency_avatar_demo.html'
output_path.write_text(html, encoding='utf-8')
print(f'Saved standalone demo to: {output_path}')
display(HTML(html))
```
